```python
# Import Numpy for calculations, matplotlib for charting, and pandas for coding
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Save path to data set in a variable
data_file = "../resources/raw/dataSet.csv"
data_file2 = "../resources/raw/counties.csv"
```

```python
# Use Pandas to read data
data_file_df = pd.read_csv(data_file)
data_file_df.head()
```

```python
#Drop NaN's from columns

data_file_df.dropna(inplace=True, subset=["sub_region_1", "sub_region_2"])
data_file_df.head(20)
```

```python
data_file2_df = pd.read_csv(data_file2)
data_file2_df.head() 
```

```python
combined_df = data_file_df[data_file_df["sub_region_1"].isin(data_file2_df["State"])]
combined_df = combined_df[combined_df["sub_region_2"].isin(data_file2_df["County"])]
#combined_df.loc[combined_df["sub_region_2"]=="Sacramento County",:]
combined_df.head()
#combined_df["sub_region_1"].value_counts()
#data_file2_df["County"].tolist()
```

```python
#load the csv
whole_df = pd.read_csv("../resources/clean/Karen_clean.csv")
whole_df
```

```python
#make the date column a datetime object
whole_df['date'] = pd.to_datetime(whole_df['date'], format="%Y/%m/%d")
whole_df.head()
```

```python
#change column names
columns_df = whole_df.rename(columns = {"sub_region_1":"State", "sub_region_2":"County", "retail_and_recreation_percent_change_from_baseline":"Retail/Rec", "grocery_and_pharmacy_percent_change_from_baseline":"Grocery/Pharmacy", "parks_percent_change_from_baseline":"Parks", "transit_stations_percent_change_from_baseline":"Transit Stations", "workplaces_percent_change_from_baseline":"Workplaces", "residential_percent_change_from_baseline":"Residential"})
columns_df.head()
```

```python
#add a new column that is the month
columns_df["month"] = columns_df["date"].map(lambda x: x.month)
#group on regions and month
grouped_df = columns_df.groupby(["State","month"]).mean()
#group by state and attributes
grouped_df = grouped_df[["Retail/Rec","Grocery/Pharmacy","Parks","Transit Stations","Workplaces","Residential"]]
grouped_df
```

```python
#replace NaN's with zero
new_grouped_df = grouped_df.fillna(0)
df = new_grouped_df.reset_index()
```

```python
#prepare information for graphing movement in indoor public spaces
df_new = df.groupby("State").mean()
stateslist1 = dict()
stateslist1["Max Retail/Rec"]=df_new.sort_values(by = "Retail/Rec").head(5).index
stateslist1["Min Retail/Rec"]=df_new.sort_values(by = "Retail/Rec").tail(5).indexstateslist2 = dict()
stateslist2 = dict()
stateslist2["Max Shopping"]=df_new.sort_values(by = "Grocery/Pharmacy").head(5).index
stateslist2["Min Shopping"]=df_new.sort_values(by = "Grocery/Pharmacy").tail(5).index
stateslist3 = dict()
stateslist3["Max Outside"]=df_new.sort_values(by = "Parks").head(5).index
stateslist3["Min Outside"]=df_new.sort_values(by = "Parks").tail(5).index
stateslist4 = dict()
stateslist4["Max Transit"]=df_new.sort_values(by = "Transit Stations").head(5).index
stateslist4["Min Transit"]=df_new.sort_values(by = "Transit Stations").tail(5).index
stateslist5 = dict()
stateslist5["Max Work"]=df_new.sort_values(by = "Workplaces").head(5).index
stateslist5["Min Work"]=df_new.sort_values(by = "Workplaces").tail(5).index
stateslist6 = dict()
stateslist6["Max Home"]=df_new.sort_values(by = "Residential").head(5).index
stateslist6["Min Home"]=df_new.sort_values(by = "Residential").tail(5).index
```

```python
#print top 5 minimums and maximums for each state showing movement in indoor public spaces
stateslist1
```

```python
#graph states with maximum movement over time in indoor spaces per month
import matplotlib as mpl
mpl.style.use('seaborn-darkgrid')
df_max_rr = [df.loc[df["State"] == x]["Retail/Rec"] for x in stateslist1["Max Retail/Rec"]]
df_max_month = [df.loc[df["State"] == x]["month"] for x in stateslist1["Max Retail/Rec"]]
fig,ax=plt.subplots()
ax.plot(df_max_month[0], df_max_rr[0],label="Hawaii")
ax.plot(df_max_month[0], df_max_rr[1],label="Louisiana")
ax.plot(df_max_month[0], df_max_rr[2],label="California")
ax.plot(df_max_month[0], df_max_rr[3],label="New York")
ax.plot(df_max_month[0], df_max_rr[4],label="New Jersey")
ax.set_title("Maximum Public Indoor Places Per Month")
ax.set_ylabel("Percent Change")
ax.set_xlabel("Month 2020")
plt.savefig("Karens_Graphs/Max_Indoor.png")
leg = ax.legend();
plt.show
```
