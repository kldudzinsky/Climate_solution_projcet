```python
xls_path = "../resources/raw/Primary_Energy_Production_by_Source.xlsx"
energy_source_monthly = pd.read_excel(xls_path, sheet_name = 0)
energy_source_yearly = pd.read_excel(xls_path, sheet_name = 1)
xls_path = "../resources/raw/Carbon_Dioxide_Emissions_From_Energy_Consumption_by_Source.xlsx"
co2_ouput_monthly = pd.read_excel(xls_path, sheet_name = 0)
co2_output_yearly = pd.read_excel(xls_path, sheet_name = 1)
#units in Quadrillion Btu
energy_source_monthly.head()
```

```python
#monthly fossil fuel production over time.
fig, ax = plt.subplots(figsize = (10,6))
ax = plt.plot(energy_source_monthly["Month"], energy_source_monthly["Total Fossil Fuels Production"])
```

```python
#Yearly fossil fuel production over time *****
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(energy_source_yearly["Annual Total"], energy_source_yearly["Total Fossil Fuels Production"])
ax.set_title("Fossil Fuel Energy Production (US)", fontsize = 22)
ax.set_ylabel("Quadrillion Btu", fontsize = 15)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.savefig("fossil_fuel_production_line.png")
plt.show()
```

```python
#Renewable energy production over time *****
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(energy_source_yearly["Annual Total"], energy_source_yearly["Total Renewable Energy Production"])
ax.set_title("Renewable Energy Production (US)", fontsize = 22)
ax.set_ylabel("Quadrillion Btu", fontsize = 15)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.savefig("renewable_energy_production_line.png")
plt.show()
```

```python
#Coal production over time
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(energy_source_yearly["Annual Total"], energy_source_yearly["Coal Production"])
ax.plot(energy_source_yearly["Annual Total"], energy_source_yearly["Natural Gas (Dry) Production"])
ax.plot(energy_source_yearly["Annual Total"], energy_source_yearly["Total Renewable Energy Production"])
ax.set_title("Dominant Fossils Fuels vs Renewable Energy (US)", fontsize = 22)
ax.set_ylabel("Quadrillion Btu", fontsize = 15)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.legend(["Coal", "Natural Gas","Total Renewable"], fontsize = 13)
plt.savefig("coal_ng_production_lines.png")
plt.show()
```

```python
#US Total Co2 emissions over time **********
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(co2_output_yearly["Annual Total"], co2_output_yearly["Total Energy CO2 Emissions"])
ax.set_title("Total Co2 Emissions (US)", fontsize = 22)
ax.set_ylabel("Million Metric Tons of Carbon Dioxide", fontsize = 15)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.savefig("total_co2_line.png")
plt.show()
```

```python
#US Co2 Emessions from coal over time
fig, ax = plt.subplots(figsize = (10,6))
ax = plt.plot(co2_output_yearly["Annual Total"],
              co2_output_yearly["Coal, Including Coal Coke Net Imports, CO2 Emissions"])
plt.xticks([x*5 + 1970 for x in range(11)])
```

# Use these pie charts

```python
#remove string for no data and replace with 0
energy_source_yearly.replace(to_replace="Not Available", value=0.0, inplace=True)
#prepare data for pie chart
droped_totals_df = energy_source_yearly.drop(columns = ["Total Primary Energy Production",
                                                        "Total Fossil Fuels Production",
                                                        "Nuclear Electric Power Production",
                                                        "Hydroelectric Power Production",
                                                        "Geothermal Energy Production",
                                                        "Solar Energy Production",
                                                        "Wind Energy Production",
                                                        "Biomass Energy Production"])
#get a series of just the first row
beginning_data = droped_totals_df.iloc[31,1:]
beginning_data = beginning_data[beginning_data > 0.0]
#get a series for just the last row
ending_data = droped_totals_df.iloc[len(droped_totals_df)-1, 1:]
ending_data = ending_data[ending_data > 0.0]
fig, ax = plt.subplots(2, figsize = (15,20))
beg_xplode = [0.0 if x != "Coal Production" else 0.1 for x in beginning_data.index]
end_xplode = [0.0 if x != "Coal Production" else 0.1 for x in ending_data.index]
ax[0].pie(beginning_data, autopct = '%1.1f%%', shadow = True,
          labels = beginning_data.index, explode = beg_xplode, 
          textprops = {'fontsize': 14})
ax[0].set_title("Total US Energy Production by Source (1970)",fontsize = 22)
ax[1].pie(ending_data, autopct = '%1.1f%%', shadow = True,
          labels = ending_data.index, explode = end_xplode,
          textprops = {'fontsize': 14})
ax[1].set_title("Total US Energy Production by Source (2019)",fontsize = 22)
plt.savefig("production_by_source_pies.png")
plt.show()
```

```python
emissions_file = "../resources/raw/emissions.xlsx"
emissions_df = pd.read_excel(emissions_file, sheet_name = 1)
emissions_df
emissions_df = emissions_df.loc[emissions_df["Code"].isin(["BIT","DFO","NG"]),:]
emissions_df.replace(["Bituminous Coal", "Distillate Fuel Oil"], ["Coal","Crude Oil"], inplace = True)
emissions_df
```

```python
fig, ax = plt.subplots()
ax.bar(emissions_df["Type"],emissions_df["Amount (Kilograms of CO2 Per Million Btu)"], 
       color = plt.cm.Paired(np.arange(len(emissions_df["Type"]))))
ax.set_title("Co2 Output by Source", fontsize = 22)
ax.set_ylabel("Kilograms of CO2 Per Million Btu", fontsize = 15)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.savefig("co2_output_bar.png")
plt.show()
```
